```python
from collections import deque

with open('sdag.txt', 'r') as file:
    lines = file.read().splitlines()

n, m = map(int, lines[0].split())
adj = [[] for _ in range(n + 1)]
in_degree = [0] * (n + 1)

for line in lines[1:m + 1]:
    u, v, w = map(int, line.split())
    adj[u].append((v, w))
    in_degree[v] += 1

queue = deque()
for i in range(1, n + 1):
    if in_degree[i] == 0:
        queue.append(i)

top_order = []
while queue:
    u = queue.popleft()
    top_order.append(u)
    for v, w in adj[u]:
        in_degree[v] -= 1
        if in_degree[v] == 0:
            queue.append(v)

inf = float('inf')
dist = [inf] * (n + 1)
dist[1] = 0

for u in top_order:
    for v, w in adj[u]:
        if dist[u] + w < dist[v]:
            dist[v] = dist[u] + w

D = []
for i in range(1, n + 1):
    if dist[i] == inf:
        D.append("x")
    else:
        D.append(str(int(dist[i])))

print(' '.join(D))
```
