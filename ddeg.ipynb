```python
with open('ddeg.txt', 'r') as file:
    lines = file.read().splitlines()

n, m = map(int, lines[0].split())
degrees = [0] * (n + 1)
adj = [[] for _ in range(n + 1)]

for line in lines[1:]:
    u, v = map(int, line.split())
    degrees[u] += 1
    degrees[v] += 1
    adj[u].append(v)
    adj[v].append(u)

D = []
for i in range(1, n + 1):
    sum_deg = sum(degrees[v] for v in adj[i])
    D.append(str(sum_deg))

print(' '.join(D))
```
