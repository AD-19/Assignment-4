bf.ipynb

```python
with open('bf.txt', 'r') as file:
    lines = file.read().splitlines()

n, m = map(int, lines[0].split())
edges = []
for line in lines[1:]:
    parts = line.split()
    if len(parts) == 3:
        u, v, w = map(int, parts)
        edges.append((u, v, w))

inf = float('inf')
dist = [inf] * (n + 1)
dist[1] = 0

for i in range(n - 1):
    for u, v, w in edges:
        if dist[u] != inf and dist[u] + w < dist[v]:
            dist[v] = dist[u] + w

D = []
for j in range(1, n + 1):
    if dist[j] == inf:
        D.append("x")
    else:
        D.append(str(int(dist[j])))

print(' '.join(D))
```

cc.ipynb

```python
with open('cc.txt', 'r') as file:
    lines = file.read().splitlines()

n, m = map(int, lines[0].split())
adj = [[] for _ in range(n + 1)]
for line in lines[1:]:
    u, v = map(int, line.split())
    adj[u].append(v)
    adj[v].append(u)

visited = [False] * (n + 1)
count = 0

for u in range(1, n + 1):
    if not visited[u]:
        count += 1
        stack = [u]
        while stack:
            node = stack.pop()
            if not visited[node]:
                visited[node] = True
                for neighbor in adj[node]:
                    if not visited[neighbor]:
                        stack.append(neighbor)

print(count)
```

bfs.ipynb

```python
from collections import deque #better than list in this case (add to end or popping made in constant time) but list also works (in o(n)?)

with open('bfs.txt', 'r') as file:
    lines = file.read().splitlines()

n, m = map(int, lines[0].split())
adj = [[] for i in range(n + 1)]

for line in lines[1:]:
    u, v = map(int, line.split())
    adj[u].append(v)

distances = [-1] * (n + 1)
distances[1] = 0
queue = deque([1])

while queue:
    current = queue.popleft()
    for neighbor in adj[current]:
        if distances[neighbor] == -1:
            distances[neighbor] = distances[current] + 1
            queue.append(neighbor)

print(' '.join(map(str, distances[1:])))
```

cte.ipynb

```python
import heapq

with open('cte.txt', 'r') as file:
    lines = file.read().splitlines()

index = 0
k = int(lines[index])
index += 1
results = []

for i in range(k):
    while index < len(lines) and lines[index].strip() == '':
        index += 1
    if index >= len(lines):
        break
    n, m = map(int, lines[index].split())
    index += 1
    adj = [[] for i in range(n + 1)]
    edges = []
    for i in range(m):
        u, v, w = map(int, lines[index].split())
        adj[u].append((v, w))
        edges.append((u, v, w))
        index += 1
    if not edges:
        results.append("-1")
        continue
    first_u, first_v, first_w = edges[0]
    dist = [float('inf')] * (n + 1)
    dist[first_v] = 0
    heap = []
    heapq.heappush(heap, (0, first_v))
    while heap:
        current_dist, u = heapq.heappop(heap)
        if current_dist > dist[u]:
            continue
        for neighbor, weight in adj[u]:
            if dist[neighbor] > dist[u] + weight:
                dist[neighbor] = dist[u] + weight
                heapq.heappush(heap, (dist[neighbor], neighbor))
    if dist[first_u] != float('inf'):
        cycle_length = first_w + dist[first_u]
        results.append(str(cycle_length))
    else:
        results.append("-1")

print(' '.join(results))
```

bip.ipynb

```python
with open('bip.txt', 'r') as file:
    lines = file.read().splitlines()

index = 0
k = int(lines[index])
index += 1
results = []

for i in range(k):
    while index < len(lines) and lines[index].strip() == '':
        index += 1
    if index >= len(lines):
        break
    n, m = map(int, lines[index].split())
    index += 1
    adj = [[] for i in range(n + 1)]
    for i in range(m):
        u, v = map(int, lines[index].split())
        adj[u].append(v)
        adj[v].append(u)
        index += 1
    color = [0] * (n + 1)
    is_bip = True
    for u in range(1, n + 1):
        if color[u] == 0:
            queue = [u]
            color[u] = 1
            while queue and is_bip:
                current = queue.pop(0)
                for neighbor in adj[current]:
                    if color[neighbor] == 0:
                        color[neighbor] = -color[current]
                        queue.append(neighbor)
                    elif color[neighbor] == color[current]:
                        is_bip = False
                        break
    if is_bip:
        results.append("1")
    else:
        results.append("-1")

print(' '.join(results))
```

dag.ipynb

```python
with open('dag.txt', 'r') as file:
    lines = file.read().splitlines()

index = 0
k = int(lines[index])
index += 1
results = []

for i in range(k):
    while index < len(lines) and lines[index].strip() == '':
        index += 1
    if index >= len(lines):
        break
    n, m = map(int, lines[index].split())
    index += 1
    adj = [[] for i in range(n + 1)]
    in_degree = [0] * (n + 1)
    for i in range(m):
        u, v = map(int, lines[index].split())
        adj[u].append(v)
        in_degree[v] += 1
        index += 1
    queue = []
    for j in range(1, n + 1):
        if in_degree[j] == 0:
            queue.append(j)
    count = 0
    while queue:
        u = queue.pop(0)
        count += 1
        for v in adj[u]:
            in_degree[v] -= 1
            if in_degree[v] == 0:
                queue.append(v)
    if count == n:
        results.append("1")
    else:
        results.append("-1")

print(' '.join(results))
```

dij.ipynb

```python
import heapq  #same as bfs, complexity o(?) better for insert and retrieve

with open('dij.txt', 'r') as file:
    lines = file.read().splitlines()

n, m = map(int, lines[0].split())
adj = [[] for i in range(n + 1)]
for line in lines[1:]:
    u, v, w = map(int, line.split())
    adj[u].append((v, w))

dist = [float('inf')] * (n + 1)
dist[1] = 0
queue = []
heapq.heappush(queue, (0, 1))

while queue:
    current_dist, u = heapq.heappop(queue)
    if current_dist > dist[u]:
        continue
    for v, w in adj[u]:
        if dist[v] > dist[u] + w:
            dist[v] = dist[u] + w
            heapq.heappush(queue, (dist[v], v))

D = []
for i in range(1, n + 1):
    if dist[i] == float('inf'):
        D.append("-1")
    else:
        D.append(str(int(dist[i])))

print(' '.join(D))
```

grph.ipynb

```python
def read_fasta(filename):
    sequences = {}
    with open(filename, 'r') as file:
        label = ""
        seq = ""
        for line in file:
            line = line.strip()
            if line.startswith('>'):
                if label:
                    sequences[label] = seq
                label = line[1:]
                seq = ""
            else:
                seq += line
        if label:
            sequences[label] = seq
    return sequences

def overlap_graph(sequences, k):
    edges = []
    for s_label, s_seq in sequences.items():
        suffix = s_seq[-k:]
        for t_label, t_seq in sequences.items():
            if s_label != t_label and t_seq.startswith(suffix):
                edges.append(f"{s_label} {t_label}")
    return edges

def main():
    sequences = read_fasta('grph.txt')
    edges = overlap_graph(sequences, 3)
    for edge in edges:
        print(edge)

main()
```

sdag.ipynb

```python
from collections import deque

with open('sdag.txt', 'r') as file:
    lines = file.read().splitlines()

n, m = map(int, lines[0].split())
adj = [[] for _ in range(n + 1)]
in_degree = [0] * (n + 1)

for line in lines[1:m + 1]:
    u, v, w = map(int, line.split())
    adj[u].append((v, w))
    in_degree[v] += 1

queue = deque()
for i in range(1, n + 1):
    if in_degree[i] == 0:
        queue.append(i)

top_order = []
while queue:
    u = queue.popleft()
    top_order.append(u)
    for v, w in adj[u]:
        in_degree[v] -= 1
        if in_degree[v] == 0:
            queue.append(v)

inf = float('inf')
dist = [inf] * (n + 1)
dist[1] = 0

for u in top_order:
    for v, w in adj[u]:
        if dist[u] + w < dist[v]:
            dist[v] = dist[u] + w

D = []
for i in range(1, n + 1):
    if dist[i] == inf:
        D.append("x")
    else:
        D.append(str(int(dist[i])))

print(' '.join(D))
```

long.ipynb

```python
def read_fasta(filename):
    with open(filename) as f:
        data = f.read().split('>')[1:]
    sequences = []
    for entry in data:
        lines = entry.strip().split('\n')
        sequences.append(''.join(lines[1:]))
    return sequences

def find_overlap(s1, s2):
    max_overlap = 0
    for i in range(1, len(s1)):
        if s2.startswith(s1[i:]):
            max_overlap = len(s1) - i
            break
    return max_overlap

def assemble_superstring(sequences):
    while len(sequences) > 1:
        max_overlap = -1
        pair_to_merge = (0, 0)
        for i in range(len(sequences)):
            for j in range(len(sequences)):
                if i != j:
                    overlap = find_overlap(sequences[i], sequences[j])
                    if overlap > max_overlap:
                        max_overlap = overlap
                        pair_to_merge = (i, j)
        i, j = pair_to_merge
        sequences[i] = sequences[i] + sequences[j][max_overlap:]
        sequences.pop(j)
    return sequences[0]

sequences = read_fasta("long.txt")
result = assemble_superstring(sequences)
print(result)
```

nwc.ipynb

```python
with open('nwc.txt', 'r') as file:
    lines = file.read().splitlines()

index = 0
k = int(lines[index])
index += 1
results = []

for i in range(k):
    while index < len(lines) and lines[index].strip() == '':
        index += 1
    if index >= len(lines):
        break
    n, m = map(int, lines[index].split())
    index += 1
    edges = []
    for i in range(m):
        u, v, w = map(int, lines[index].split())
        edges.append((u, v, w))
        index += 1
    dist = [0] * (n + 1)
    for i in range(n):
        updated = False
        for u, v, w in edges:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                updated = True
        if not updated:
            break
    else:
        results.append("1")
        continue
    has_negative = False
    for u, v, w in edges:
        if dist[u] + w < dist[v]:
            has_negative = True
            break
    if has_negative:
        results.append("1")
    else:
        results.append("-1")

print(' '.join(results))
```
